# Data Extraction

```python
SIZE = 200
start0, end0 = 0, 25
dfs = [fetch_data(from_=0, size=25)]
for i in range(0,5):
    start_i = end0+1 + SIZE * i
    print(start_i)
    temp_df = fetch_data(from_=start_i, size=SIZE)
    dfs.append(temp_df)
    sleep(1)
```

# Data Transformation

```python
df = pd.concat(dfs)
df = df.reset_index(drop=True)
df = df.rename(
    columns={
        'HitHighlightedSummary': 'content',
        'Source.From.EmailAddress.Address': 'sender',
        'Source.LastDeliveryTime': 'mail_timestamp_utc',
        }
    )
df = df.loc[df['sender'] == 'enviodigital@bancochile.cl']
df = df.drop(columns=['Source.Preview', 'Rank'])
df['currency'] = np.where(df['raw_amount'].str.contains(','), 'USD', 'CLP')
df['amount'] = (
    df['raw_amount']
    .str.replace('$', '',regex=False)
    .str.replace('.', '',regex=False)
    .str.replace(',', '.',regex=False)
    .astype(float)
    )
df['reason'] = df['content'].str.extract(r'(\d{4}) en (.*?) el')[1]
df['transaction_timestamp_local'] = df['content'].str.extract(r' el (.*?)\.')

df['mail_timestamp_utc'] = pd.to_datetime(df['mail_timestamp_utc'])
df['transaction_timestamp_local'] = pd.to_datetime(df['transaction_timestamp_local'], format='%d/%m/%Y %H:%M')
df = df[['Id', 'mail_timestamp_utc', 'transaction_timestamp_local', 'sender', 'currency', 'amount', 'reason', 'content']]
```

# Send to Supabase

```python
try:
    db_url = f"postgresql://{SUPABASE_USER}:{SUPABASE_PASSWORD}@{SUPABASE_HOST}:{SUPABASE_PORT}/{SUPABASE_DBNAME}"
    engine = create_engine(db_url)

    df.drop_duplicates(inplace=True)
    df.to_sql(
        'transactions',
        con=engine,
        if_exists='replace',
        index=False
    )
    
except Exception as e:
    print(f"An error occurred: {e}")
```
